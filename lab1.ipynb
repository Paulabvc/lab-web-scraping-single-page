```python
def scrape_hot100():
    '''
    Scrapes the 100 hottest songs from billboard.com/charts/hot-100
    and stores the respective singer name and songtitle to a csv file.
    
    Return:
        On success we return a pandas dataframe containing the singer and song.
        On error (Failed to fetch data from URL) None is returned.
    '''

    url = 'https://www.billboard.com/charts/hot-100.'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    if response.status_code != 200:
        print("! Failed to download page content")
        return None

    songs   = []
    singers = []

    for x in soup.select('div.o-chart-results-list-row-container'):
        li = x.select('ul > li')[3]
        songs.append(li.h3.get_text().strip())
        singers.append(li.span.get_text().strip())
    
    df = pd.DataFrame({'Singer': singers,
                       'Song' : songs})
    df.to_csv('hot_100.csv', index=False)

    return df
```
